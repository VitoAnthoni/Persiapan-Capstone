```
#Import masing-masing library
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```
#Bagian ini menampilkan dataframe awal sebelum menjalani pengolahan
res = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
soup = BeautifulSoup(res.content,'lxml')
table = soup.find_all('table')[0] 
df = pd.read_html(str(table))[0]
df
```

```
#Bertujuan mengdrop baris yang memiliki nilai Borough Not assigned
#For dropping rows that has Not Assigned Value in Borough
drop_index = df[df['Borough']=='Not assigned'].index
df.drop(drop_index , inplace = True)
df
```

```
df = df.reset_index()
```

```
df = df.drop(columns=['index'])
```

```
df
```

```
for i in range(210):
  Neigh = df.iloc[i,2]
  Borough = df.iloc[i,1]
  if Neigh == 'Not assigned' :
    df.iat[i,2] = Borough
```

```
df.head(n=20)
```

```
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
```

```
df
```

```
df = df.sort_values('Postcode')
```

```
df_lat_ling = pd.read_csv("http://cocl.us/Geospatial_data")
```

```
lat= df_lat_ling['Latitude']
lon= df_lat_ling['Longitude']
```

```
df['Latitude'] = lat
```

```
df['Longitude'] = lon
```

```
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
lat = location.latitude
lng = location.longitude

print('The geograpical coordinates of Toronto are {}, {}.'.format(lat, lng))
```

```
!conda install -c conda-forge folium=0.5.0 --yes
import folium # map rendering library
print('Folium libraries imported.')
```

```
# create map of Toronto using latitude and longitude values
map_toronto = folium.Map(location=[lat, lng], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='maroon',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.5,
        parse_html=False).add_to(map_toronto)
```

```
map_toronto
```

```
from folium import plugins

# let's start again with a clean map of Toronto
map_toronto = folium.Map(location = [lat, lng], zoom_start = 10)

# instantiate a marker cluster object for the postcodes
postcodes = plugins.MarkerCluster().add_to(map_toronto)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, postcode in zip(df['Latitude'], df['Longitude'], df['Postcode']):
    label = 'lat-long: {}<br>Postcode: {}'.format(location[1], postcode)
    label = folium.Popup(label, parse_html=False)
    folium.Marker(
        location=[lat, lng],
        icon=folium.Icon(color='green', icon='ok-sign'),
        popup=label
    ).add_to(postcodes)
```
